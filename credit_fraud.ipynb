```python
def displots(df, cols, cond=None, color='b', title='', add_norm=False):
    if cond is None:
        cond = [True]*df.shape[0]
    
    fig, axes = plt.subplots(1, len(cols), figsize=(20,5))
#     if not (type(axes) is list): axes = [axes]
    fig.suptitle(title)
    for ax, col in zip(axes, cols):
        ser = df.loc[cond, col]
        sns.histplot(ser.values, ax=ax, kde=True, stat="density", linewidth=0, color=color)
        if add_norm:
            normx = np.linspace(ser.min(), ser.max(), 100)
            ax.plot(normx, norm.pdf(normx, *norm.fit(ser.values)), lw=1, color='black')
        ax.set_title(f'Distribution of Transaction {col.capitalize()}', fontsize=14)
#         ax.set_xlim([min(val), max(val)])
```

## Downsampling

```python
is_fraud = (df['Class']==1)
fraud_df = df.loc[is_fraud]
non_fraud_df = df.loc[~is_fraud].sample(492)

bdf = pd.concat([fraud_df, non_fraud_df]).sample(frac=1, random_state=0)

bdf['Class'].value_counts().plot.pie(autopct='%1.1f%%')
```

```python
def boxplots(df, cols, cond=None, title=''):
    if cond is None:
        cond = [True]*df.shape[0]
    
    fig, axes = plt.subplots(1, len(cols), figsize=(20,5))
    fig.suptitle(title, fontsize=18)
    for ax, col in zip(axes, cols):
        sns.boxplot(x="Class", y=col, data=df, palette=['green', 'red'], ax=ax)
        ax.set_title(f'{col} vs Class Correlation')
```

```python
def remove_fraud(df, cols, cond=None, threshold=1.5):
    
    for col in cols:
        fraud = df.loc[cond, col]
        q25 = np.percentile(fraud, 25) 
        q75 = np.percentile(fraud, 75)
        iqr = q75 - q25
        
        lower = q25 - iqr*1.5         
        upper = q75 + iqr*1.5 
        is_outlier = ((df[col] < lower) | (df[col] > upper))
        print(col, "Outliers")
        display(is_outlier.value_counts())
        df = df[~is_outlier]
    return df
```

```python
bdf_new = remove_fraud(bdf, corr_ser.index[:3], cond=(bdf['Class']==1))
```

```python
bdf_new = remove_fraud(bdf_new, corr_ser.index[::-1][:3], cond=(bdf_new['Class']==1))
```

## Modelling

```python
from sklearn.model_selection import train_test_split

X = bdf_new.drop('Class', axis=1)
y = bdf_new['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier

classifiers = {
    "LogisiticRegression": LogisticRegression(),
    "KNearest": KNeighborsClassifier(),
    "SupportVectorClassifier": SVC(),
    "DecisionTreeClassifier": DecisionTreeClassifier(),
    "RandomForestClassifier": RandomForestClassifier(),
    "GradientBoostingClassifier": GradientBoostingClassifier()
}
```

```python
from sklearn.model_selection import cross_val_score

for classifier_name, classifier in classifiers.items():
    classifier.fit(X_train, y_train)
    score = cross_val_score(classifier, X_train, y_train, cv=5)
    print(f"{classifier_name} training score {score.mean()*100:.2f}")
```

```python
from sklearn.model_selection import GridSearchCV

def gridsearch(classifier_name, classifier, classifier_params):
    grid_classifier = GridSearchCV(classifier, classifier_params)
    grid_classifier.fit(X_train, y_train)

    classifier = grid_classifier.best_estimator_
    classifier_score = cross_val_score(classifier, X_train, y_train, cv=5)
    print(f'{classifier_name} CV Score: {classifier_score.mean()*100:.2f}%')

    return classifier
```

```python
for classifier_name, (classifier, classifier_params) in classifiers.items():
    classifier_best_estimator = gridsearch(classifier_name, classifier, classifier_params)
    classifiers[classifier_name] = classifier_best_estimator
```
