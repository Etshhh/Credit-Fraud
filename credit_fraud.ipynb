```python
def boxplots(df, cols, cond=None, title=''):
    if cond is None:
        cond = [True]*df.shape[0]
    
    fig, axes = plt.subplots(1, len(cols), figsize=(20,5))
    fig.suptitle(title, fontsize=18)
    for ax, col in zip(axes, cols):
        sns.boxplot(x="Class", y=col, data=df, palette=['green', 'red'], ax=ax)
        ax.set_title(f'{col} vs Class Correlation')
```

```python
def remove_fraud(df, cols, cond=None, threshold=1.5):
    
    for col in cols:
        fraud = df.loc[cond, col]
        q25 = np.percentile(fraud, 25) 
        q75 = np.percentile(fraud, 75)
        iqr = q75 - q25
        
        lower = q25 - iqr*1.5         
        upper = q75 + iqr*1.5 
        is_outlier = ((df[col] < lower) | (df[col] > upper))
        print(col, "Outliers")
        display(is_outlier.value_counts())
        df = df[~is_outlier]
    return df
```
